```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import warnings
warnings.filterwarnings('ignore')

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.inspection import permutation_importance
import shap

plt.rcParams['axes.unicode_minus'] = False

print("="*80)
print("   FUEL CELL CURRENT GENERATION ESTIMATOR + ANALYSIS SUITE")
print("="*80)

# ====================== CLEAN LABEL FUNCTION ======================
def clean_label(col):
    col = col.replace('ｰ', '°')
    col = col.replace('(C)', '(°C)')
    col = col.replace('(ｰC)', '(°C)')
    return col

# ====================== LOAD DATA ======================
file_name = 'Fuel cell data.csv'
df = pd.read_csv(file_name)

df.columns = df.columns.str.replace('ｲ', '2')

target = 'I (A)'
exclude_cols = ['Time (h)', 'I (A)', 'J (A/cm2)', 'Utot (V)']
features = [col for col in df.columns if col not in exclude_cols]

X = df[features]
y = df[target]

# ====================== TRAIN MODEL ======================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

model = RandomForestRegressor(n_estimators=150, random_state=42, n_jobs=-1)
model.fit(X_train, y_train)

joblib.dump((model, features), 'fuel_cell_current_model.pkl')

# ====================== PREDICTION ======================
y_pred = model.predict(X_test)

# ====================== METRICS ======================
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("\n📊 MODEL PERFORMANCE:")
print(f"MAE  : {mae:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"R2   : {r2:.4f}")
print(f"MAPE : {mape:.2f}%")

# ====================== INTERACTIVE INPUT ======================
ranges = {col: (df[col].min(), df[col].max()) for col in features}

print("\nEnter input values:\n")
input_data = {}

for f in features:
    mn, mx = ranges[f]
    val = input(f"{clean_label(f)} [{mn:.3f}-{mx:.3f}]: ")
    input_data[f] = float(val) if val else (mn + mx) / 2

loaded_model, feature_list = joblib.load('fuel_cell_current_model.pkl')
input_df = pd.DataFrame([input_data])[feature_list]

predicted_current = loaded_model.predict(input_df)[0]

print("\n" + "="*80)
print(f"Predicted I (A): {predicted_current:.4f}")
print("="*80)

# ====================== RESIDUALS ======================
residuals = y_test - y_pred

# ====================== 1. Actual vs Predicted ======================
plt.figure()
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Current (A)")
plt.ylabel("Predicted Current (A)")
plt.title("Actual vs Predicted Current")
plt.show()

# ====================== 2. Residual Plot ======================
plt.figure()
plt.scatter(y_pred, residuals)
plt.axhline(0)
plt.xlabel("Predicted Current (A)")
plt.ylabel("Residual Error (A)")
plt.title("Residual Plot")
plt.show()

# ====================== 3. Error Distribution ======================
plt.figure()
plt.hist(residuals, bins=30)
plt.title("Error Distribution (Residuals)")
plt.xlabel("Error (A)")
plt.ylabel("Frequency")
plt.show()

# ====================== 4. Time Series ======================
plt.figure()
plt.plot(y_test.values, label="Actual")
plt.plot(y_pred, label="Predicted")
plt.legend()
plt.title("Time Series Prediction of Current")
plt.show()

# ====================== 5. Feature Importance ======================
imp = model.feature_importances_
plt.figure()
plt.barh(features, imp)
plt.title("Feature Importance (Random Forest)")
plt.show()

# ====================== 6. Permutation Importance ======================
perm = permutation_importance(model, X_test, y_test)
plt.figure()
plt.barh(features, perm.importances_mean)
plt.title("Permutation Importance")
plt.show()

# ====================== 7. Density Heatmap ======================
bins = 20
heat, _, _ = np.histogram2d(y_test, y_pred, bins=bins)

plt.figure()
sns.heatmap(heat.T, cmap="viridis", xticklabels=False, yticklabels=False)
plt.title("Actual vs Predicted Density Map")
plt.xlabel("Actual Bins")
plt.ylabel("Predicted Bins")
plt.show()

# ====================== 8. Error vs Actual ======================
plt.figure()
plt.scatter(y_test, residuals)
plt.xlabel("Actual Current (A)")
plt.ylabel("Error (A)")
plt.title("Error vs Actual Current")
plt.show()

# ====================== 9. Error vs Predicted ======================
plt.figure()
plt.scatter(y_pred, residuals)
plt.xlabel("Predicted Current (A)")
plt.ylabel("Error (A)")
plt.title("Error vs Predicted Current")
plt.show()

# ====================== 10. SHAP ======================
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test[:200])
shap.summary_plot(shap_values, X_test[:200], show=True)

# ====================== 11. Sensitivity Plots ======================
for i in range(2):
    plt.figure()
    plt.scatter(X_test.iloc[:, i], y_pred)
    plt.xlabel(clean_label(features[i]))
    plt.ylabel("Predicted Current (A)")
    plt.title(f"Sensitivity Analysis: {clean_label(features[i])}")
    plt.show()

# ====================== 12. 3D Colored Prediction Surface ======================
from mpl_toolkits.mplot3d import Axes3D

f1, f2 = features[0], features[1]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(
    X_test[f1],
    X_test[f2],
    y_pred,
    c=y_pred,
    cmap='viridis',
    s=20
)

ax.set_xlabel(clean_label(f1))
ax.set_ylabel(clean_label(f2))
ax.set_zlabel("Predicted Current (A)")
ax.set_title("3D Prediction Surface (Colored)")

plt.colorbar(sc)
plt.show()

# ====================== 13. 3D Error Surface ======================
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(
    X_test[f1],
    X_test[f2],
    residuals,
    c=residuals,
    cmap='coolwarm',
    s=20
)

ax.set_xlabel(clean_label(f1))
ax.set_ylabel(clean_label(f2))
ax.set_zlabel("Residual Error (A)")
ax.set_title("3D Error Surface (Colored)")

plt.colorbar(sc)
plt.show()

# ====================== 14. Correlation Matrix ======================
plt.figure()
sns.heatmap(df[features + [target]].corr(), cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# ====================== 15. Predicted Distribution ======================
plt.figure()
plt.hist(y_pred, bins=30)
plt.title("Predicted Current Distribution")
plt.xlabel("Current (A)")
plt.ylabel("Frequency")
plt.show()

# ====================== 16. Actual Distribution ======================
plt.figure()
plt.hist(y_test, bins=30)
plt.title("Distribution of Actual Current (A)")
plt.xlabel("Current (A)")
plt.ylabel("Frequency")
plt.show()

# ====================== 17. Absolute Error Trend ======================
plt.figure()
plt.plot(np.abs(residuals))
plt.title("Absolute Error Trend Across Test Samples")
plt.xlabel("Sample Index (Time-ordered)")
plt.ylabel("|Error| (A)")
plt.show()

# ====================== 18. Cumulative Error ======================
plt.figure()
plt.plot(np.cumsum(np.abs(residuals)))
plt.title("Cumulative Absolute Error in Prediction")
plt.xlabel("Sample Index (Time-ordered)")
plt.ylabel("Cumulative Error (A)")
plt.show()
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import os
import warnings
warnings.filterwarnings('ignore')

# GPU Acceleration Setup (using XGBoost with GPU support for significant speedup)
# Note: Requires NVIDIA GPU with CUDA installed and 'xgboost' package with GPU support.
# If GPU is not available, it will fall back to CPU.
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.inspection import permutation_importance
import shap

print("="*80)
print(" FUEL CELL CURRENT GENERATION ESTIMATOR + ANALYSIS SUITE (GPU-ACCELERATED)")
print("="*80)

# ====================== LOAD DATA ======================
file_name = 'Fuel cell data.csv'
df = pd.read_csv(file_name)

# Fix encoding issues for proper temperature symbols
df.columns = df.columns.str.replace('ｲ', '2')
df.columns = df.columns.str.replace('ｰ', '°')  # Proper degree symbol for temperatures

target = 'I (A)'
exclude_cols = ['Time (h)', 'I (A)', 'J (A/cm2)', 'Utot (V)']
features = [col for col in df.columns if col not in exclude_cols]

X = df[features]
y = df[target]

# ====================== TRAIN MODEL (GPU-ACCELERATED) ======================
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# XGBoost with GPU support for faster training and prediction
model = XGBRegressor(
    n_estimators=300,          # Increased for better performance
    learning_rate=0.05,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    tree_method='gpu_hist',    # GPU acceleration (use 'hist' if no GPU)
    gpu_id=0,                  # GPU device ID
    predictor='gpu_predictor'
)

print("Training model with GPU acceleration...")
model.fit(X_train, y_train)

# Save model and features
joblib.dump((model, features), 'fuel_cell_current_model.pkl')
print("Model saved successfully.")

# ====================== PREDICTION ======================
y_pred = model.predict(X_test)

# ====================== ERROR METRICS ======================
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("\n📊 MODEL PERFORMANCE:")
print(f"MAE  : {mae:.4f} A")
print(f"RMSE : {rmse:.4f} A")
print(f"R²   : {r2:.4f}")
print(f"MAPE : {mape:.2f}%")

# ====================== INTERACTIVE INPUT ======================
ranges = {col: (df[col].min(), df[col].max()) for col in features}
print("\nEnter input values for prediction:\n")

input_data = {}
for f in features:
    mn, mx = ranges[f]
    # Display proper °C label for temperature parameters
    display_name = f.replace('°C', ' (°C)') if '°C' in f else f
    val = input(f"{display_name} [{mn:.3f} - {mx:.3f}]: ")
    input_data[f] = float(val) if val.strip() else (mn + mx) / 2

# Load model and predict
loaded_model, feature_list = joblib.load('fuel_cell_current_model.pkl')
input_df = pd.DataFrame([input_data])[feature_list]
predicted_current = loaded_model.predict(input_df)[0]

print("\n" + "="*80)
print(f"Predicted I (A): {predicted_current:.4f}")
print("="*80)

# ====================== ENHANCED PLOTS (with proper labels and 3D visualizations) ======================
plt.rcParams['figure.figsize'] = (10, 6)

# 1 Actual vs Predicted
plt.figure()
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("Actual Current (A)")
plt.ylabel("Predicted Current (A)")
plt.title("Actual vs Predicted Current")
plt.grid(True)
plt.show()

# 2 Residual plot
residuals = y_test - y_pred
plt.figure()
plt.scatter(y_pred, residuals, alpha=0.6)
plt.axhline(0, color='r', linestyle='--')
plt.xlabel("Predicted Current (A)")
plt.ylabel("Residuals (A)")
plt.title("Residual Plot")
plt.grid(True)
plt.show()

# 3 Histogram of errors
plt.figure()
plt.hist(residuals, bins=30, edgecolor='black')
plt.xlabel("Residual (A)")
plt.ylabel("Frequency")
plt.title("Error Distribution")
plt.grid(True)
plt.show()

# 4 Time series comparison
plt.figure()
plt.plot(y_test.values, label="Actual", alpha=0.8)
plt.plot(y_pred, label="Predicted", alpha=0.8)
plt.xlabel("Test Sample Index")
plt.ylabel("Current (A)")
plt.title("Time Series Prediction Comparison")
plt.legend()
plt.grid(True)
plt.show()

# 5 Feature importance
imp = model.feature_importances_
plt.figure()
plt.barh(features, imp)
plt.xlabel("Importance Score")
plt.title("Feature Importance (XGBoost)")
plt.grid(True)
plt.show()

# 6 Permutation importance
perm = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42)
plt.figure()
plt.barh(features, perm.importances_mean)
plt.xlabel("Mean Importance")
plt.title("Permutation Importance")
plt.grid(True)
plt.show()

# 7 Binned heatmap
bins = 20
heat, xedges, yedges = np.histogram2d(y_test, y_pred, bins=bins)
plt.figure()
sns.heatmap(heat.T, cmap="viridis", cbar_kws={'label': 'Count'})
plt.xlabel("Actual Current (A)")
plt.ylabel("Predicted Current (A)")
plt.title("Binned Actual vs Predicted Heatmap")
plt.show()

# 8 Error vs Actual Current
plt.figure()
plt.scatter(y_test, residuals, alpha=0.6)
plt.axhline(0, color='r', linestyle='--')
plt.xlabel("Actual Current (A)")
plt.ylabel("Residual (A)")
plt.title("Error vs Actual Current")
plt.grid(True)
plt.show()

# 9 Error vs Predicted Current
plt.figure()
plt.scatter(y_pred, residuals, alpha=0.6)
plt.axhline(0, color='r', linestyle='--')
plt.xlabel("Predicted Current (A)")
plt.ylabel("Residual (A)")
plt.title("Error vs Predicted Current")
plt.grid(True)
plt.show()

# 10 SHAP Summary (using first 200 samples for speed)
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test.iloc[:200])
shap.summary_plot(shap_values, X_test.iloc[:200], plot_type="bar", show=True)
shap.summary_plot(shap_values, X_test.iloc[:200], show=True)

# 11 Sensitivity (Partial Dependence style)
for i in range(min(3, len(features))):
    plt.figure()
    plt.scatter(X_test.iloc[:, i], y_pred, alpha=0.6)
    plt.xlabel(features[i])
    plt.ylabel("Predicted Current (A)")
    plt.title(f"Sensitivity Analysis: {features[i]}")
    plt.grid(True)
    plt.show()

# 12 Enhanced 3D Scatter Plot (Response Surface)
from mpl_toolkits.mplot3d import Axes3D
f1, f2 = features[0], features[1]  # First two features (typically temperatures)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_test[f1], X_test[f2], y_pred, c=y_pred, cmap='viridis', alpha=0.7)
ax.set_xlabel(f1)
ax.set_ylabel(f2)
ax.set_zlabel("Predicted Current (A)")
plt.title("3D Response Surface (Predicted Current)")
fig.colorbar(scatter, ax=ax, label='Predicted Current (A)')
plt.show()

# 13 Enhanced 3D Color Scatter - Error Surface
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_test[f1], X_test[f2], residuals, c=residuals, cmap='coolwarm', alpha=0.7)
ax.set_xlabel(f1)
ax.set_ylabel(f2)
ax.set_zlabel("Residual (A)")
plt.title("3D Error Surface")
fig.colorbar(scatter, ax=ax, label='Residual (A)')
plt.show()

# 14 Correlation Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df[features + [target]].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Features and Target")
plt.show()

# 15 Predicted Distribution
plt.figure()
plt.hist(y_pred, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel("Predicted Current (A)")
plt.ylabel("Frequency")
plt.title("Predicted Current Distribution")
plt.grid(True)
plt.show()

# 16 Actual Distribution
plt.figure()
plt.hist(y_test, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel("Actual Current (A)")
plt.ylabel("Frequency")
plt.title("Actual Current Distribution")
plt.grid(True)
plt.show()

# 17 Absolute Error Trend (over test samples)
plt.figure()
plt.plot(np.abs(residuals), marker='o', linestyle='-', alpha=0.7)
plt.xlabel("Test Sample Index")
plt.ylabel("Absolute Error (A)")
plt.title("Absolute Error Trend Across Test Samples")
plt.grid(True)
plt.show()

# 18 Cumulative Absolute Error
plt.figure()
plt.plot(np.cumsum(np.abs(residuals)), color='darkred')
plt.xlabel("Test Sample Index")
plt.ylabel("Cumulative Absolute Error (A)")
plt.title("Cumulative Absolute Error Trend")
plt.grid(True)
plt.show()

print("\nAll visualizations completed successfully.")
```
